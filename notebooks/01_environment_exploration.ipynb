```python
import sys
import os

# Determine the project root dynamically
def get_project_root():
    current_dir = os.getcwd()
    
    # Check if we're in the notebooks directory
    if os.path.basename(current_dir) == 'notebooks':
        return os.path.dirname(current_dir)
    
    return current_dir

# Add project root to Python path
project_root = get_project_root()
sys.path.insert(0, project_root)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.environment import ScarseGridEnv

# Print project root for verification
print(f"Project Root: {project_root}")

# Rest of your notebook code
env = ScarseGridEnv(grid_size=5, seed_value=42)  # Note the seed_value parameter

# Reset environment multiple times
plt.figure(figsize=(15, 5))
for i in range(3):
    state, _ = env.reset()
    plt.subplot(1, 3, i+1)
    sns.heatmap(state, annot=True, cmap='YlGnBu')
    plt.title(f'Environment State {i+1}')

plt.tight_layout()
plt.show()
```

```python
# Agent Movement Simulation
def simulate_agent_movement(env, num_steps=5):
    state, _ = env.reset()
    print("Initial Agent Positions:", env.agent_positions)
    
    movement_history = [state]
    
    for _ in range(num_steps):
        # Random actions for both agents
        actions = [np.random.randint(4), np.random.randint(4)]
        state, rewards, done, _, _ = env.step(actions)
        movement_history.append(state)
        
        print(f"Actions: {actions}, Rewards: {rewards}, New Positions: {env.agent_positions}")
    
    return movement_history

movement_history = simulate_agent_movement(env)

# Visualize movement progression
plt.figure(figsize=(15, 3))
for i, state in enumerate(movement_history):
    plt.subplot(1, len(movement_history), i+1)
    sns.heatmap(state, annot=True, cmap='YlGnBu')
    plt.title(f'Step {i}')

plt.tight_layout()
plt.show()
```
